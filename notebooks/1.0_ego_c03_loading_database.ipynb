## **IMPORTS**

```python
import sqlite3
import pandas as pd
from sqlalchemy import create_engine
```

```python
# **===================== Import dataset =====================**
data = pd.read_csv('../datasets/products_HM_cleaned.csv')
```

```python
data.sample(5)
```

```python
data.dtypes
```

```python
query_showroom_schema = """
    CREATE TABLE showroom (
        product_id         INTEGER,
        product_category   TEXT,
        product_name       TEXT,
        product_price      REAL,
        scrapy_datetime    TEXT,
        style_code         INTEGER,
        color_id           INTEGER,
        color_name         TEXT,
        fit                TEXT,
        size_model         REAL,
        size_number        TEXT,
        cotton             REAL,
        polyester          REAL,
        elastane           REAL,
        elasterell         REAL    
    )
"""
```

```python
# Connect to DataBase
conn = sqlite3.connect('hm_db.sqlite3')
cursor = conn.execute(query_showroom_schema)
conn.commit()
conn.close()
```

```python
conn = create_engine('sqlite:///hm_db.sqlite3', echo = False)
```

```python
# Insert data
data.to_sql('showroom', con = conn, if_exists = 'append', index = False)
```

```python
query = """
    SELECT *
    FROM showroom
"""
df = pd.read_sql_query(query, conn)
df.head()
```
